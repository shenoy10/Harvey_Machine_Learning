```python
import numpy as np
import pandas as pd
import os
```

```python
chunksize = 1000
count=0;
path = '/corral-repl/utexas/Trump-Tweets/Harvey/unt_data/'
#path = '../harvey-twitter/data/'

#read the frames in chunks and just keep the text and retweet count
frames = []
for chunk in pd.read_json(path + "harvey_twitter_dataset/02_archive_only/HurricaneHarvey.json", chunksize=chunksize,lines=True):
    f=pd.DataFrame(chunk)
    g=f[['text', 'retweet_count']]
    frames.append(g)
    
    count+=1000
    if count % 100000 == 0:
        print(count)

#concatenate all the pruned dataframes into 1 large one
general = pd.concat(frames)

print(general.shape)
general.head()
```

Now we just go ahead and save the dataframe as a csv file, so the really long pruning process here does not have to be run again.

```python
general.to_csv("/corral-repl/utexas/Trump-Tweets/Harvey/twitter_text.csv")
#general.to_csv("data/twitter_text.csv")
```

```python
import numpy as np
import pandas as pd
```

```python
chunksize = 1000
count=0;


generalFrame=pd.DataFrame({'text' : pd.Series() })
for chunk in pd.read_json("/corral-repl/utexas/Trump-Tweets/Harvey/unt_data/harvey_twitter_dataset/02_archive_only/HurricaneHarvey.json", chunksize=chunksize,lines=True):
    f=pd.DataFrame(chunk)
    g=f['text']
    generalFrame=generalFrame.append(pd.DataFrame(g))
    count+=1000
    #print(generalFrame)
    print(count)
```

```python
generalFrame.to_csv("/corral-repl/utexas/Trump-Tweets/Harvey/twitter_text.csv")
```
